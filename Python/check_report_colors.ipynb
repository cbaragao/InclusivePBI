```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from colorblind import colorblind
from selenium import webdriver
from time import sleep

#open page, wait about 20 secs - already signed into PBI service
driver = webdriver.Edge()
driver.get('URL')
sleep(20)

driver.get_screenshot_as_file("screenshot.png")
driver.quit()
```

```python
#load image
img = cv2.imread("screenshot.png")
img = img[..., ::-1]
```

```python
# simulate protanopia
simulated_img = colorblind.simulate_colorblindness(img, colorblind_type='protanopia')
plt.imshow(simulated_img.astype(np.uint8))
plt.show()
```

```python
# correct using daltonization
daltonized_img = colorblind.daltonize_correct(img, colorblind_type='protanopia')
plt.imshow(daltonized_img.astype(np.uint8))
plt.show()
```

```python
# correct using hsv correction
hsv_img = colorblind.hsv_color_correct(img, colorblind_type='protanopia')
plt.imshow(hsv_img.astype(np.uint8))
plt.show()
```

```python
# correct using CBFS
cbfs_img = colorblind.cbfs_correct(img, closeness=70)
plt.imshow(cbfs_img.astype(np.uint8))
plt.show()
```

```python
# correct using LAB color correction
lab_img = colorblind.lab_correct(img)
plt.imshow(lab_img.astype(np.uint8))
plt.show()
```
